# **개념 알아가기**
## **⭐ 데이터 분석을 위한 중요 개념**

데이터 분석을 위한 중요 개념들을 몇 가지 짚고 넘어가겠습니다.
- Index vs Feature vs Target
- Numeric Data vs Categorical Data
- Regression vs Classification

</br>

## **Feature vs Target**

머신러닝 데이터셋은 <font style="color:#FF9999; font-weight : bold; font-size : 25px;">인덱스(index)</font>, <font style="color:#FFCC00; font-weight : bold; font-size : 25px;">특성(feature)</font>, <font style="color:#99CCFF; font-weight : bold; font-size : 25px;">대상(target)</font>으로 이루어져 있습니다.
</br> 인덱스(index): 식별자 변수
* 즉, 샘플 이름/ID
특성(feature): 예측을 위해 활용되는 데이터
* 독립변수 X라고도 함
대상(target): 예측하고자 하는 대상
* 종속변수 y라고도 함

![image](https://ifh.cc/g/yq1JKb.jpg)

</br>

## **Numerical Data vs Categorical Data**
</br><font style="color : #666666; font-weight : bold; font-size : 25px;">Numerical Data</font>: 수치형 자료, 숫자로 측정할 수 있는 자료
* ex) 기온, 풍속, 체중 등

<font style="color : #666666; font-weight : bold; font-size : 25px;">Categorical Data</font>: 범주형 자료, 숫자로 측정할 수 없는 자료
* ex) 날씨, 혈액형, 성별 등

</br>

## **Regression vs Classification**

<font style="color : #6495ED ; font-weight : bold ; font-size : 25px;">"target의 형태"</font>에 따라 분류(classification)문제인지 회귀(regression)문제인지 나뉩니다.

</br> <font style="color : #666666; font-weight : bold; font-size : 25px;">Classification</font>: 분류
</br> 범주형 데이터(categorical data) 예측에 사용
* ex) 손글씨 이미지(feature)를 보고 숫자(target)를 예측하는 문제

</br> <font style="color : #666666; font-weight : bold; font-size : 25px;">Regression</font>: 회귀
</br> 수치형 데이터(numerical data) 예측에 사용
* ex) 지난 10일 간의 주식 가격을 보고 내일 주식의 등락을 예측하는 문제
</br>

## **결측치 처리**

정형 데이터를 다루다 보면,  column에 비어있는 값들이 존재합니다. </br>이 값들을 결측치 라고 합니다. </br>결측값을 제거하고 관측값만 분석하게 되면 일반적으로 통계적 편향이 생겨 정확한 결과를 얻기 힘들어집니다. </br>이에 결측값을 추정한 대체값을 넣어 분석하게 되는데, 이를 <font style="color:#99CCFF; font-weight : bold; font-size : 25px;">결측값처리(imputation)</font> 라 합니다.

</br>결측치를 처리하는 방법은 다양합니다.

- 특정 새로운 값(상수, 평균값, 최빈값 등)으로 대체하는 방식
- Hot deck imputation : 다른 변수에서 비슷한 값을 갖는 데이터 중 하나를 랜덤 샘플링 하여 대체하는 방식
- Cold deck imputation : 다른 변수에서 비슷한 값을 갖는 데이터 중에서 하나를 골라(규칙이 존재) 그 값으로 대체하는 방식
- Regression imputation : 결측치가 존재하지 않는 변수를 feature로 삼고, 결측치를 채우고자 하는 변수를 target으로 삼아 regression task를 진행하는 방식
- Stochastic regression imputation : regression 방법에 random residual value를 더해서 결측치의 최종 예측값으로 대체하는 방식
- 보간법, 보외법 : 같은 대상으로부터 얻은 다른 관측치로부터 결측치 부분을 추정하는 방식


</br>이 방법들은 <font style="font-weight : bold; font-size : 25px">Single Imputation</font>과  <font style="font-weight : bold; font-size : 25px">Multiple Imputation</font>로 분류할 수 있습니다.

</br>

> 1️⃣ Single Imputation 

대체값으로 <font style="color:#99CCFF; font-weight : bold; font-size : 25px;">하나의 값</font>을 선정합니다. 예시로는 hot deck, 평균값 대체, regression imputation 등이 있습니다. </br>하나의 값으로 하게 된다면 결측치가 랜덤하게 생성된 경우가 아니라면 추정 편향이 생길 수 있습니다.  </br>그러다 보면 대체값이 관측값보다 더 높은 신뢰도를 가진다고 평가될 수도 있습니다. 

</br>
이를 막기위해 나온것이 Multiple Imputation입니다.

</br>

> 2️⃣ Multiple Imputation 

대체값으로 <font style="color:#99CCFF; font-weight : bold; font-size : 25px;">여러개의 값</font>을 선정하는 방식입니다. 여러개의 데이터 셋을 만들어 평가하므로 관측값은 변함이 없어 </br>동일값들이 많이나온 것처럼 평가되는 것에 비해
대체값은 만들 때마다 추가해주는 residual로 인한 변동 때문에 </br>관측값보다 상대적으로 신뢰도가 낮게 평가됩니다.

</br>

# **모델링(Modeling)**
분석용 데이터를 이용한 가설 설정을 통하여 통계모형을 만들거나 </br>기계학습(Machine Learning)을 이용한 데이터의 분류, 예측, 군집 등의 기능을</br> 수행하는 모형을 만드는 과정입니다.

</br>기계학습은 <font style = "font-weight:bold;">지도학습(Supervised Learning)</font>, <font style = "font-weight:bold;">비지도학습(Un-supervised Learning)</font>, <font style = "font-weight:bold;">강화학습(reinforcement learning)</font>으로 나뉘어 다양한 알고리즘을 적용할 수 있습니다. 

</br>모델링을 효과적으로 진행하기 위해서는 모델링 전에 데이터셋을 Training과 Testing으로 <font style="color:#99CCFF; font-weight : bold; font-size : 25px;">"분할"</font>함으로써 </br><font style = "font-weight:bold;">모형의 Overfitting을 방지하거나 모형의 Generalization에 이용</font>됩니다.
</br>기계학습 등을 이용한 데이터 모델링은 Training 데이터를 활용하여 분류, 예측, 군집 등의 모형을 만들어 가동중인 운영 시스템에 적용합니다. </br>또한, 필요시 비정형 데이터 분석결과를 통합적으로 활용하여 프로젝트 목적에 맞는 통합 모델링을 수행합니다.

</br><font style="color:#99CCFF; font-weight : bold; font-size : 25px;">Supervised Learning</font>의 경우, 알고리즘에 입력 데이터 세트가 제공되며 </br>특정 출력 세트에 부합하도록 보상이 주어지거나 최적화됩니다. </br>예를 들어 Supervised Learning은 이미지 인식에 광범위하게 활용되는데, 이 경우 분류라는 기법을 활용합니다. </br>Supervised Learning은 인구 성장이나 건강 지표와 같은 인구통계 예측에도 쓰이며, </br>이 경우 회귀라는 기법을 활용합니다.

</br><font style="color:#99CCFF; font-weight : bold; font-size : 25px;">Un-supervised Learning</font>의 경우, 알고리즘에 입력 데이터 세트는 제공하지만 </br>특정 출력으로 보상되거나 최적화되지는 않고, 그 대신 공통된 특징에 따라 개체를 그룹으로 묶도록 교육합니다. </br>예컨대 온라인 매장의 추천 엔진은 Un-supervised Learning, 그중에서도 특히 클러스터링이라는 기법을 주로 이용합니다.

</br><font style="color:#99CCFF; font-weight : bold; font-size : 25px;">reinforcement learning</font>의 경우, 알고리즘이 수많은 시행착오 실험을 통해 자체적으로 훈련하도록 설정됩니다. </br>이 알고리즘이 훈련 데이터에 의존하는 것이 아니라 주변 환경과 계속 상호작용을 주고받을 때 일어납니다. </br>reinforcement learning의 가장 보편적인 예시가 바로 자율 주행입니다.



</br>머신 러닝 모델에도 여러 가지 종류가 있습니다.
</br>
><font style = "font-weight:bold; font-size : 25px">1️⃣ Supervised Learning</font>  

- Logistic Regression : 어느 입력이 특정 그룹에 속하는지 아닌지 판단하는 데 쓰입니다.
- SVM : 서포트 벡터 머신(Support Vector Machine)은 n차원 공간에서 각 개체의 좌표를 만들고, 초평면을 사용해 여러 개체를 공통된 특징에 따라 그룹으로 묶습니다.
- Naive Bayes: 변수 간에 비종속성이 성립한다고 가정하고 확률을 사용해 기능에 따라 개체를 분류하는 알고리즘입니다.
- Decision Tree : 분류자의 일종으로, 트리의 잎과 노드를 횡단 이동하여 주어진 입력이 어느 카테고리에 속하는지 판단하는 데 쓰입니다.
- linear regression: 관심사와 입력 변수 사이의 관계를 파악하고, 입력 변수 값에 따라 그 값을 예측하는 데 쓰입니다.
- KNN: k Nearest Neighbors 기법은 가장 가까운 개체를 하나의 데이터 세트로 그룹화한 다음 개체 중에서 가장 자주 나타나거나 평균적인 특징을 찾아내는 것 위주입니다.
- Random Forest : 무작위 데이터 하위 집합에서 가져온 수많은 결정 트리를 모은 컬렉션입니다. 따라서 결정 트리 하나만 사용할 때보다 좀 더 정확한 예측을 내놓을 수 있는 트리 조합이 생깁니다.
- Boosting: Gradient Boosting Machine, XGBoost와 LightGBM 등의 Boosting은 앙상블 학습을 사용합니다. 이 알고리즘은 여러 알고리즘(예: 결정 트리)에서 가져온 예측을 조합하되, 이전 알고리즘에서 발생한 오류를 고려합니다.

</br>

><font style = "font-weight:bold; font-size : 25px">2️⃣ Un-supervised Learning</font>

- K-Means: 여러 개체의 유사상을 찾아 이를 K개의 서로 다른 클러스터로 그룹화합니다.
- hierachical clustering: 클러스터 수를 지정할 필요 없이 중첩된 클러스터 트리를 구축합니다.

## **선형 회귀(Linear Regression)**

Linear Regression는 널리 사용되는 대표적인 회귀 알고리즘입니다. </br>선형 회귀는 종속 변수 y와 하나 이상의 독립 변수 x와의 선형 상관관계를 모델링하는 기법입니다. </br>만약, 독립 변수 x가 1개라면 단순 선형 회귀라고 하고, 2개 이상이면 다중 선형 회귀라고 합니다. 

## 단순 선형 회귀(Simple Linear Regression)

단순 선형 회귀는 y=Wx+b의 식으로 나타냅니다. </br>머신러닝에서는 독립 변수 x에 곱해지는 W값을 가중치(weight), 상수항에 해당하는 b를 편향(bias)이라고 부릅니다. 
</br>따라서 단순 선형 회귀 모델을 훈련하는 것은 적절한 W와 b값을 찾는 것입니다. </br>그래프의 형태는 직선으로 나타납니다. 

## 다중 선형 회귀 (Multiple Linear Regression)

다중 선형 회귀는 y=W1x1+W2x2+...+Wnxn+b의 식으로 나타납니다. </br>여러 독립 변수에 의해 영향을 받는 경우입니다. 
</br>만약, 2개의 독립 변수면 그래프는 평면으로 나타날 것입니다.
